```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from keras.preprocessing.text import Tokenizer
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
df = pd.read_csv('successData.csv')
# df = df.iloc[:500, :]
```

```python
df['companyName'] = df['companyName'].astype(str)
df['actor1'] = df['actor1'].astype(str)
df['actor2'] = df['actor2'].astype(str)
df['actor3'] = df['actor3'].astype(str)
# df['budget'] = df['budget'].astype(int)
df['genre1'] = df['genre1'].astype(str)
df['genre2'] = df['genre2'].astype(str)
# df['totalMin'] = df['totalMin'].astype(float)
# df['ageGroup'] = df['ageGroup'].astype(str)
df['writer1'] = df['writer1'].astype(str)
df['writer2'] = df['writer2'].astype(str)
df['director'] = df['director'].astype(str)
df['rating'] = df['rating'].astype(float)
```

```python
df.head()
```

```python
# df = df["companyName"] + df["director"] + df["writer1"] + df["writer2"] + df["actor1"] + df["actor2"] + df["actor3"] + df["rating"] 
df = df[["companyName", "director","writer1","writer2","actor1","actor2",'actor3',"genre1","genre2","rating"]]
df.head()
```

```python
df.shape
```

```python
listData = []
for index, row in df.iterrows():
    listData.append([row[0], row[1], row[2], row[3],
                    row[4], row[5], row[6], row[7], row[8]])
listData
```

```python
t = Tokenizer()
fit_text = listData
t.fit_on_texts(fit_text)
test_text = [["O2 Filmes"], ["Avi Arad Productions"],
             ["Niki Caro"]]
sequences = t.texts_to_sequences(test_text)
sequences
```

```python
df.head()
```

```python
for i, row in df.iterrows():
    ifor_val = row[0:9]
    dataList = []
    # print(type(row))
    # print(list(ifor_val))
    for ind in list(ifor_val):
        dataList.append([ind])

    # print(dataList)
    sequences = t.texts_to_sequences(dataList)
    sequences.append([row[9]])
    # print(sequences)
    newDataList = []
    for x in sequences:
        newDataList.append(x[0])

    # print(newDataList)

    df.loc[i, ['companyName', 'director','writer1', 'writer2', 'actor1', 'actor2', 'actor3', "genre1", "genre2", 'rating']] = newDataList

df.head()
```

```python
df = df.dropna(how='any',axis=0)
df.head()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0)

X_train
```

```python
# RFRregressor = RandomForestRegressor(
#     n_estimators=750, random_state=0,max_features=0.75)
RFRregressor = RandomForestRegressor(
    n_estimators= 1000, random_state=0, max_features=1, n_jobs=-1)
RFRregressor.fit(X_train, y_train)
y_pred = RFRregressor.predict(X_test)
r2_score = RFRregressor.score(X_test, y_test)
np.set_printoptions(precision=2)
print(r2_score*100, '%')
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```
